```python
import pandas as pd
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
df.describe()
```

```python
list1 =['mainroad','guestroom','basement',"hotwaterheating",'airconditioning','prefarea']
df[list1] = df[list1].replace({'yes':1,'no':0})
df.head()
```

```python
df.tail()
```

```python
list2 =['furnishingstatus']
df[list2] = df[list2].replace({'furnished':1,'semi-furnished':0.5,"unfurnished":0})
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
list3 = ['price','area']
df[list]=std.fit_transform(df[list])
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


#Scatter plot of price vs. area
plt.figure(figsize=(8, 6))
sns.scatterplot(x='area', y='price', data=df)
plt.title('Price vs. Area')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
# Distribution of price
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], kde=True, color='red')
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Pairplot for selected features
sns.pairplot(df[['price', 'area', 'bedrooms', 'bathrooms']])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('price',axis=1)
y = df['price']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
acc = lr.score(x_test,y_test)
acc
```

```python

plt.figure(figsize=(4,6))
plt.bar(['Accuracy'], [acc],color = 'green')
plt.ylim(0, 1)  # Set y-axis limit to 0-1 for accuracy
plt.ylabel('Accuracy')
plt.title('Model Accuracy')
plt.show()
```
